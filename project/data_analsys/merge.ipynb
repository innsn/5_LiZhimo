```python
import pandas as pd
```

```python
data = pd.read_csv("D:/计算机大三上/大数据分析实践/project/CN-Reanalysis201301/201301/CN-Reanalysis-daily-2013010100.csv")
data['data'] = '13-1-1'
for month in range(1,13):
    for day in range(32):
        try:
            with open("D:/计算机大三上/大数据分析实践/project/CN-Reanalysis201301/201301/CN-Reanalysis-daily-201301{}00.csv".format(day),'r',encoding='utf-8') as file:
                temp_data=pd.read_csv(file)
                temp_data['data'] = '13-{}-{}'.format(month,day)
        except FileNotFoundError:
            continue
        data = pd.concat([data,temp_data])
```
