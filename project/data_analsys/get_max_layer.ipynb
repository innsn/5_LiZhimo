```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("D:/homework/Bigdata_Analysis_Practice/201301/CN-Reanalysis-daily-2013010100.csv")
df_2 = pd.read_csv("D:/homework/Bigdata_Analysis_Practice/data/latlon.csv", encoding='GBK')
for date in range(2, 32):
    with open(
            "D:/homework/Bigdata_Analysis_Practice/201301/CN-Reanalysis-daily-201301{}00.csv".format(
                    str(date).zfill(2)), encoding='utf-8') as f:
        df_temp = pd.read_csv(f)
        df = pd.concat([df, df_temp])
```

```python
df = df.drop(df.columns[-1], axis=1)
df = pd.merge(df,df_2,how='left',on=[' lat',' lon'])
df
```

```python
#df.to_csv('df.csv',index='false')
last_column = df.iloc[:, [-2,-3]]
df = df.drop(columns=[' O3(微克每立方米)', ' U(m/s)', ' V(m/s)', ' TEMP(K)',
       ' RH(%)', ' PSFC(Pa)', ' lat',' lon', '省', '市'])
df
```

```python
last_column
```

```python
df[' CO(毫克每立方米)'] = df[' CO(毫克每立方米)'] * 1000
total = df.iloc[:,0:5].sum(axis=1)
df = df.iloc[:, 0:5].div(total, axis=0)
```

```python
df = pd.concat([df, last_column], axis=1)
df
```

```python
df.to_csv("dfshengshi.csv")
```

```python
group = df.groupby(['省', '市'])
```

```python
g_mean = group.mean()
g_mean
```

```python
g_var = group.var()
g_var
```

```python
max_layer = g_mean + g_var / g_mean
max_layer
```

```python
min_layer = g_mean - g_var / g_mean
min_layer
```
