```python
import json
import pandas as pd
import requests

data = pd.read_csv("D:/计算机大三上/大数据分析实践/project/CN-Reanalysis201301/201301/CN-Reanalysis-daily-2013010100.csv")

key = "c7317dd02d5b2c06ff11d30362bbbf38"
# key = "d7e17d69de210e72e240edd0241ed225"
# key = "0a8fc3d2f20ad60edec568674b33bc82"
total = []
city = []
province = []
district = []
```

```python

city = []
province = []
district = []
index1 = 20800
index2 = 22500
for item1,item2 in data[[' lat',' lon']][index1:index2].values:
    # print(item1)
    location = str(item2) + ','+ str(item1)
    url = "https://restapi.amap.com/v3/geocode/regeo?output&location=" + location + "&key="+key+"&radius=1000&extensions=all"
    r = requests.get(url)

    # print(r.text)
    t = r.text
    js = json.loads(t)
    # locations = {'lat':[],'lon':[],'city':[],'province':[],'district':[]}
    # locations['lat'].append(item1)
    # locations['lon'].append(item2)
    # locations['city'].append(js['regeocode']['addressComponent']['district'])
    # locations['province'].append(js['regeocode']['addressComponent']['province'])
    # locations['district'].append(js['regeocode']['addressComponent']['district'])
    city.append(js['regeocode']['addressComponent']['city'])
    province.append(js['regeocode']['addressComponent']['province'])
    district.append(js['regeocode']['addressComponent']['district'])
```

```python
city = pd.Series(city)
province = pd.Series(province)
district = pd.Series(district)

merged_series = pd.concat([data[[' lat',' lon']][index1:index2].reset_index(), city, province, district], axis=1, ignore_index=True)
merged_series
```

```python
merged_series.to_csv("../location/{}-{}.csv".format(index1,index2-1))
```
