```python
import pandas as pd
import os
```

```python
directory = '../location/location/'

file_names = os.listdir(directory)

files = [os.path.join(directory, file) for file in file_names]

df = pd.DataFrame()
for file in files:
    try:
        with open(file) as f:
            data = pd.read_csv(f)
    except:
        with open(file,encoding='gbk') as f:
            data = pd.read_csv(f)
    data = data.drop(data.columns[0],axis=1)
    if ' lat' not in data.columns:
        data = data.drop(data.columns[0],axis=1)
        data = data.rename(columns={'1':' lat', '2':' lon', '3': '0', '4': '1', '5': '2'})
    df = pd.concat([df, data])
    print(file,data.head(1).values)
```
